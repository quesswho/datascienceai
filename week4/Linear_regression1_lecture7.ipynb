```python
from sklearn.linear_model import LinearRegression            # Import library

# Generate random data
rng= np.random.RandomState(3) # define rng, Create a random number generator with a fixed seed
x= 10*rng.rand(50)            # Generate 50 random values for x, scaled to the range [0, 10]
y= 2*x-5 + rng.randn(50)      # Generate y values based on a linear equation with some noise

# Create and fit the linear regression model
model = LinearRegression()      # Create an instance of the LinearRegression model
model.fit(x[:, np.newaxis], y)  # model.fit(x,y) ; Fit the model using x (reshaped to 2D) and y


# Create data for predictions
xfit=np.array([0,10])         # Define x values for which we want to predict y
yfit= model.predict (xfit[:,np.newaxis]) # Predict y values based on the fitted model

# Plotting
plt.scatter(x, y, label='Data Points')
plt.plot(xfit, yfit, color='purple', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Example')
plt.show()
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
# Generate random data
rng = np.random.RandomState(3)  # Create a random number generator with a fixed seed
x = 10 * rng.rand(50)           # Generate 50 random x values, scaled to the range [0, 10]
y = 2 * x - 5 + rng.randn(50)   # Generate corresponding y values with some noise

# Create and fit the linear regression model
model = LinearRegression()      # Create an instance of LinearRegression
model.fit(x[:, np.newaxis], y)  # Fit the model using x (reshaped to 2D) and y

# Create data for predictions
xfit = np.array([0, 10])  # Define x values for which we want to predict y
yfit = model.predict(xfit[:, np.newaxis])  # Predict y values based on the fitted model

# Calculate predicted y values for the original data points
y_predicted = model.predict(x[:, np.newaxis])

# Calculate residuals (difference between actual y and predicted y)
residuals = y - y_predicted

# Plot data points and the regression line
plt.scatter(x, y, label='Data Points')
plt.plot(xfit, yfit, color='green', label='Regression Line')

# Plot the residuals (vertical lines from actual y to predicted y)
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y_predicted[i]], color='red', linestyle='--')

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression with Residuals')

# Display the plot
plt.show()

# Print the residuals for reference (optional)
print("Residuals:", residuals)
```
